```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib import cm

param1 = np.linspace(5, 90, 18)
param2 = np.linspace(5, 90, 18)

Y, X = np.meshgrid(param2, param1)
```

# TOP VIEW

# solidity

```python
os.chdir('D:/theoretical_waterlily_data/csv/sil_x0-y0')

df_top = pd.read_csv('nympha-ell_silhouette_x0-y0_top_20180711part.csv')
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set_aspect('equal')
levels = [0.46, 0.52, 0.58, 0.64, 0.70, 0.76, 0.82, 0.88, 0.94, 1.00]

sld_top = df_top['solidity'].values.reshape([18,18])
im1 = ax1.contourf(X, Y, sld_top, levels, cmap=cm.viridis)
ax1.set_xlabel(r'$x_\mathrm{0}$')
ax1.set_ylabel(r'$y_\mathrm{0}$')

fig1.colorbar(im1)
plt.show()
```
