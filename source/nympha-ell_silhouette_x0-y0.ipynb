```python
# top view

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
import cv2

val = pd.DataFrame(columns = ('name','solidity','convexity'))

for name in glob.glob('D:/theoretical_waterlily_data/csv/x0-y0/*.csv'):
    #座標データのファイルを開く
    df = pd.read_csv(name, header = None)
    data = np.array(df)
    head, tail = os.path.split(name)
    filename, ext = os.path.splitext(tail)
    imagename = filename + '_top' + ".png"

    #解析する画像ファイルがないとき、生成する
    if os.path.isfile(imagename) == False:
        os.chdir('D:/theoretical_waterlily_data/png/sil_x0-y0') # 20180711フォルダ名変更 > sil_x0-y0_top
        plt.figure(figsize = (6.4, 4.8))
        plt.gca().set_aspect('equal')
        plt.scatter(data[:, 0], data[:, 1], c = 'black')
        plt.xlim([-150, 150])
        plt.ylim([-150, 150])
        plt.axis('off')
        plt.box('off')
        plt.savefig(imagename)
    #画像の前処理
    im = cv2.imread(imagename)
    im2 = cv2.bitwise_not(im)
    imgray = cv2.cvtColor(im2,cv2.COLOR_BGR2GRAY)
    ret,thresh = cv2.threshold(imgray,127,255,0)
    image, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    #凸包の計算
    draw_cont = cv2.drawContours(im2, contours, -1, (0,255,0), 3)
    hull = cv2.convexHull(contours[0])
    draw_cont2 = cv2.drawContours(im2, [hull], -1, (255,0,0), 3)
    #物体と凸包の面積計算
    area = cv2.contourArea(contours[0])
    area2 = cv2.contourArea(hull)
    #物体と凸包の周長計算
    perimeter = cv2.arcLength(contours[0],True)
    perimeter2 = cv2.arcLength(hull,True)
    #データフレームにまとめる
    add_val = pd.DataFrame([[tail, area/area2, perimeter2/perimeter]], columns = ('name','solidity','convexity'))
    val = val.append(add_val, ignore_index=True)
    #ファイルを閉じる
    cv2.destroyAllWindows()
    plt.close()
    print(imagename)
    os.chdir('D:/theoretical_waterlily_data/csv/x0-y0')

#結果の表示と記録
print(val)
val.to_csv("nympha-ell_silhouette_x0-y0_top.csv") 
# "D:/theoretical_waterlily_data/csv/sil_x0-y0" へ移動 20180711
# "nympha-ell_silhouette_x0-y0_top_20180711.csv" へファイル名変更 20180711
# x0 <= 90, y <= 90 までのデータだけを残したファイルを作成　"nympha-ell_silhouette_x0-y0_top_20180711part.csv"　20180711

print("completed")
```
