```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import os
import glob
import matplotlib.cm as cm

lim = 55
reso = 90
db = 10

#directory of output files of "nympha-ell_default.py"'
csv_dir = 'C:/Users/shiryu/Documents/GitHub/theoretical-morphological-model-of-water-lily/data/sample_theoretical_model'
csv_files = csv_dir + '/*.csv'

#directory of output folder of this note
png_dir = csv_dir

for name in glob.glob(csv_files):
    df = pd.read_csv(name, header = None)
    data = np.array(df)
    head, tail = os.path.split(name)
    filename, ext = os.path.splitext(tail)
    imagename = filename + '_3d' + ".png"
    
    os.chdir(png_dir)
    #making fig
    #fig = plt.figure(figsize = (6.4, 4.8))
    fig = plt.figure(figsize = plt.figaspect((lim + db) / (2*lim)))
    ax = fig.gca(projection='3d')
    
    #NaN check
    if data[0,0] != data[0,0]:
        continue
    
    for i in range(0, 102, 1):
        start = i*reso*reso
        end = (i+1)*reso*reso
        subdata = data[start:end, :].reshape((90,90,3))
        ax.plot_surface(subdata[:,:,0], subdata[:,:,1], subdata[:,:,2], antialiased=True, color = "crimson")
    
    #adjusting image aspect
    ax.set_xlim3d(-lim, lim)
    ax.set_ylim3d(-lim, lim)
    ax.set_zlim3d(-db, lim)
        
    plt.axis('off')
    plt.box('off')
        
    plt.savefig(imagename)
    print(imagename)
    os.chdir(csv_dir)

print("completed")
```
