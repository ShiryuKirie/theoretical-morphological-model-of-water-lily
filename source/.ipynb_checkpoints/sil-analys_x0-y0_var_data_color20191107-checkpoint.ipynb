```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib import cm
from matplotlib.colors import ListedColormap

param1 = np.linspace(0.0, 9.0, 20)
param2 = np.linspace(0.0, 9.0, 20)

Y, X = np.meshgrid(param2, param1)
```

```python
os.chdir('/GitHub/theoretical-morphological-model-of-water-lily/data')
data2 = pd.read_csv('linear-model_tepal_20190611.csv', header = 'infer')
```

```python
os.chdir('/GitHub/theoretical-morphological-model-of-water-lily/data')
df = pd.read_csv("linear-model_tepal_20190611.csv")
norm = df.loc[:, "diameter"]/2
sy = -1*df.loc[:, "coef_maj"]/norm
sx = -1*df.loc[:, "coef_min"]/norm
y0dia = df.loc[:, "intercept_maj"]/norm
y0 = y0dia/20
x0dia = df.loc[:, "intercept_min"]/norm
x0 = x0dia/20

data = pd.DataFrame({"sy":sy, "sx":sx, "y0":y0, "x0":x0, "shape":df["shape"]})
```

```python
binBig = cm.get_cmap('binary', 512)
newcmp = ListedColormap(binBig(np.linspace(0.25, 0.75, 256)))
```

# TOP VIEW
## Please set the directory which includes output CSVs of "nympha-ell_silhouette_x0-y0.ipynb"

```python
os.chdir('D:/theoretical_waterlily_data/csv/sil_x0-y0_var')
df_top = pd.read_csv('nympha-ell_silhouette_x0-y0_var_top.csv')
```

# solidity

```python
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set_aspect('equal')
levels = [0.40, 0.45, 0.50, 0.55, 0.60, 0.65, 0.70, 0.75, 0.80, 0.85, 0.90, 0.95, 1.00]

sld_top = df_top['solidity'].values.reshape([20,20])
im1 = ax1.contourf(X, Y, sld_top, levels, cmap=newcmp)

for i in range(len(df)):
    if df['shape'][i] == "cup":
        ax1.scatter(x0[i], y0[i], c='red', marker ='v')
    elif df['shape'][i] == "unknown":
        ax1.scatter(x0[i], y0[i], c='blue', marker='x')
    elif df['shape'][i] == "other":
        ax1.scatter(x0[i], y0[i], c='green',marker = 'o')
    elif df['shape'][i] == "stellate":
        ax1.scatter(x0[i], y0[i], c='purple',marker = 'D')
    else:
        print("error in " + i)

ax1.set_xlabel(r'$x_\mathrm{0}$')
ax1.set_ylabel(r'$y_\mathrm{0}$')

fig1.colorbar(im1)
plt.show()
```

# convexity

```python
fig2 = plt.figure()
ax2 = fig2.add_subplot(111)
ax2.set_aspect('equal')

cnv_top = df_top['convexity'].values.reshape([20,20])
im2 = ax2.contourf(X, Y, cnv_top,  cmap=newcmp)

for i in range(len(df)):
    if df['shape'][i] == "cup":
        ax2.scatter(x0[i], y0[i], c='red', marker ='v')
    elif df['shape'][i] == "unknown":
        ax2.scatter(x0[i], y0[i], c='blue', marker='x')
    elif df['shape'][i] == "other":
        ax2.scatter(x0[i], y0[i], c='green',marker = 'o')
    elif df['shape'][i] == "stellate":
        ax2.scatter(x0[i], y0[i], c='purple',marker = 'D')
    else:
        print("error in " + i)


ax2.set_xlabel(r'$x_\mathrm{0}$')
ax2.set_ylabel(r'$y_\mathrm{0}$')

fig2.colorbar(im2)
plt.show()
```
