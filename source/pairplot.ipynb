```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

#directory of "linear-model_tepal_20190611.csv"
dir = "here"

os.chdir(dir)
sns.set(style="darkgrid")
df = pd.read_csv("linear-model_tepal_20190611.csv")

norm = df.loc[:, "diameter"]/2

sy = -1*df.loc[:, "coef_maj"]/norm
sx = -1*df.loc[:, "coef_min"]/norm
y0dia = df.loc[:, "intercept_maj"]/norm
y0 = y0dia/20
x0dia = df.loc[:, "intercept_min"]/norm
x0 = x0dia/20

#data = pd.concat([sy,sx,y0,x0], axis = 1)
data = pd.DataFrame({"sy":sy, "sx":sx, "y0":y0, "x0":x0, "shape":df["shape"]})

#print(data)

corr_mat = data.corr(method='pearson')
print(corr_mat)

g = sns.PairGrid(data, hue = "shape", palette="Set1", hue_kws={"marker": ["v", "x", "o","D"]})
g = g.map_upper(plt.scatter, s=20)
g = g.map_lower(sns.kdeplot, cmap="binary", linewidth=1)
g = g.map_diag(plt.hist, histtype="step", linewidth = 3, alpha = 1)
g = g.add_legend()

#g.axes[0,0].set_xlim((0,3.5))
#g.axes[0,1].set_xlim((0,3.5))
#g.axes[0,2].set_xlim((0.5,9))
#g.axes[0,3].set_xlim((0.5,9))
#g.axes[0,0].set_ylim((0,3.5))
#g.axes[1,0].set_ylim((0,3.5))
#g.axes[2,0].set_ylim((0.5,9))
#g.axes[3,0].set_ylim((0.5,9))

plt.show()
```
