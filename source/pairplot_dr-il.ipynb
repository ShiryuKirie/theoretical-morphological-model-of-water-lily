```python
import os
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

os.chdir("D:/theoretical_waterlily_data/csv/measure_estimation")
sns.set(style="darkgrid")
df = pd.read_csv("linear-model_tepal_extend_20180710.csv")
#'il1_Maj', 'il1_Min', 'dr1_Maj', 'dr1_Min'
data = df.iloc[:, 8:]

corr_mat = data.corr(method='pearson')
print(corr_mat)

g = sns.PairGrid(data)
g = g.map_upper(plt.scatter, color = "black")
g = g.map_lower(sns.kdeplot, color = "black")
g = g.map_diag(plt.hist, color = "black")

plt.show()
```
