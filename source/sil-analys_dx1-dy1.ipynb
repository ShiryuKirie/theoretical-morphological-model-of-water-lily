# decreasing rate

```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
from matplotlib import cm

param1 = np.linspace(0, 0.05, 26)
param2 = np.linspace(0, 0.05, 26)

Y, X = np.meshgrid(param2, param1)
```

# TOP VIEW

# solidity

```python
os.chdir('D:/theoretical_waterlily_data/csv/sil_dx1-dy1')

df_top = pd.read_csv('nympha-ell_silhouette_dx1-dy1_top_20180711.csv')
fig1 = plt.figure()
ax1 = fig1.add_subplot(111)
ax1.set_aspect('equal')

sld_top = df_top['solidity'].values.reshape([26,26])
im1 = ax1.contourf(X, Y, sld_top, cmap=cm.viridis)
ax1.set_xlabel(r'$dx_\mathrm{1}$')
ax1.set_ylabel(r'$dy_\mathrm{1}$')

fig1.colorbar(im1)
plt.show()
```
